```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
# Đọc file dữ liệu 
df1 = pd.read_csv('file_resut_2.csv')
df1.head()
```

```python
# plot biểu đồ 
df1['waterlevel'].plot(kind='line', figsize = (12, 8))
```

```python
df1.plot.bar(figsize = (12, 8))
```

```python
df1.plot.scatter(x='datetime', y ='waterlevel', c='waterlevel', cmap='rainbow', figsize = (12, 8))
```

```python
df1['waterlevel'].plot.box(figsize = (12, 8))
```
